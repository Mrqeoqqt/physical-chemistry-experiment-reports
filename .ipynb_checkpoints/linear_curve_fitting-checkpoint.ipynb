```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
X = [18.2, 18.2, 18.5, 20.8, 21.8, 23.6, 25.0, 26.4, 27.7, 29.2, 29.6]
Y = [38.5, 27.4, 22.0, 10.6, 7.12, 4.50, 2.55, 1.95, 1.29, 0.75, 0.51]

T1 = [142.17, 86.66, 43.61, 90.59, 53.87]
T2 = [427.13, 270.78, 178.52, 195.89, 133.70]
t = [30, 35, 40, 45, 50]
```

```python
Y = np.array(np.log(1/(np.array(T2)-np.array(T1))))
X = 1/(np.array(t)+273.15)
```

```python
#直线拟合
def linear_func(X, A, B):
    return A*X + B

result = optimize.curve_fit(linear_func, X, Y)
print(result)
```

```python
A, B = result[0]
x1 = np.arange(0.00305, 0.0034, 10**(-5))
y1 = A*x1 + B
```

```python
## plot
plt.figure(figsize=(10, 8))
plt.scatter(X, Y, alpha = 0.75, label = 'experimental')
plt.plot(x1, y1, "-r", label = 'fitting curve')
plt.legend()
plt.xlim([0.003, 0.0035])
plt.xlabel(r'$\frac{1}{T}/K^{-1}$', fontsize = 20)
plt.ylabel(r'$\ln(\frac{1}{t})/\lns^{-1}$', fontsize = 20)
plt.savefig(r'C:\Users\xym\Desktop\experiment.png', dpi = 600)
plt.show()
```

```python
## R^2
def R_squared(Y, Y_hat):
    y_bar = np.mean(Y)
    SSR = np.sum(np.square(Y_hat - y_bar))
    SST = np.sum(np.square(Y - y_bar))
    return SSR/SST
```

```python
Y_hat = A*X+B
print(R_squared(Y, Y_hat))
```
