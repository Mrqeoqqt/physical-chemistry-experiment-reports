```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
X = np.array([0.0521, 
0.0974, 
0.1489, 
0.1976, 
0.2960, 
])
n = np.array([1.4244 ,
1.4215 ,
1.4179 ,
1.4146 ,
1.4081 ,
])

e = np.array([2.21 ,
2.38 ,
2.57 ,
2.76 ,
3.18 ,
])

rho = np.array([0.7818 ,
0.7849 ,
0.7895 ,
0.7939 ,
0.8040 ,
])

```

```python
#直线拟合
def linear_func(X, A, B):
    return A*X + B

result = optimize.curve_fit(linear_func, X, e)
print(result)
```

```python
A, B = result[0]
x1 = np.arange(0, 0.35, 0.001)
y1 = A*x1 + B
print(A, B)
print(A/B)
```

```python
## plot
plt.figure(figsize=(10, 8))
plt.scatter(X, e, alpha = 0.75, label = 'experimental')
plt.plot(x1, y1, "-r", label = 'fitting curve')
plt.legend()
plt.xlim([min(x1), max(x1)])
plt.xlabel(r'$x_{2}$', fontsize = 20)
plt.ylabel(r'$\epsilon_{r}$', fontsize = 20)
plt.savefig(r'C:\Users\xym\Desktop\jiedian.png', dpi = 600)
plt.show()
```

```python
## R^2
def R_squared(Y, Y_hat):
    y_bar = np.mean(Y)
    SSR = np.sum(np.square(Y_hat - y_bar))
    SST = np.sum(np.square(Y - y_bar))
    return SSR/SST
```
