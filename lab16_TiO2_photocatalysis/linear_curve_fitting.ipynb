```python
## R^2
def R_squared(Y, Y_hat):
    y_bar = np.mean(Y)
    SSR = np.sum(np.square(Y_hat - y_bar))
    SST = np.sum(np.square(Y - y_bar))
    return SSR/SST
```

```python
Y_hat = result4[0][0]*T4+result4[0][1]
print(R_squared(np.log(C4), Y_hat))
```
